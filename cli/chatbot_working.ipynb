```python
# ğŸ¯ 5. ë°©ë²• 2: Jupyterì—ì„œ subprocessë¡œ CLI ì‹¤í–‰
import subprocess
import os

def run_cli_command(command_args, background=False):
    """CLI ëª…ë ¹ì–´ë¥¼ subprocessë¡œ ì‹¤í–‰"""
    try:
        cmd = [
            'bash', '-c', 
            f'cd {project_root} && source venv/bin/activate && python cli/chatbot.py {command_args}'
        ]
        
        if background:
            # ë°±ê·¸ë¼ìš´ë“œ ì‹¤í–‰ (ëŒ€í™”í˜•ì´ ì•„ë‹Œ ëª…ë ¹ì–´ìš©)
            result = subprocess.run(cmd, capture_output=True, text=True, timeout=10)
            return result.stdout, result.stderr, result.returncode
        else:
            # ì¸í„°ë™í‹°ë¸Œ ì‹¤í–‰
            subprocess.run(cmd)
            
    except subprocess.TimeoutExpired:
        print("â° ëª…ë ¹ì–´ ì‹¤í–‰ ì‹œê°„ì´ ì´ˆê³¼ë˜ì—ˆìŠµë‹ˆë‹¤.")
    except Exception as e:
        print(f"âŒ ëª…ë ¹ì–´ ì‹¤í–‰ ì˜¤ë¥˜: {e}")

print("ğŸ”§ CLI ëª…ë ¹ì–´ ì‹¤í–‰ í•¨ìˆ˜ ì •ì˜ ì™„ë£Œ!")
print("\nì‚¬ìš© ì˜ˆì‹œ:")
print("  run_cli_command('--config config/card_issuance_chatbot.json --validate-only', background=True)")
print("  run_cli_command('--config config/card_issuance_chatbot.json --verbose')")
```
